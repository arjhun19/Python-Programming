```python
def sleep_in(weekday, vacation):
  if weekday == False:
    return True
  elif weekday == True and vacation == True:
    return True
  else:
    return False

sleep_in(True,True)
sleep_in(True,False)
sleep_in(False,False)
```

```python
sleep_in(True,False)
```

```python
sleep_in(True,True)
```

```python
def sum_double(a, b):
  if a==b and b==a:
    c=(a+b)*2
    return c
  else:
    c=a+b
    return c
```

```python
sum_double(3,3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def deterministic(m, c):
    x=np.random.random(100)
    y=m*x+c
    plt.scatter(x,y)

def statical (m , c):
    x=np.random.random(100)
    y=m*x+c+np.random.random(100)
    plt.scatter(x,y)
```

```python
deterministic(2,8)
```

```python
statical(2,8)
```
